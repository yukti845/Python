```
from bs4 import BeautifulSoup
import requests
url = "https://www.iiitd.ac.in/people/faculty"
```

```
page = requests.get(url)
print(page.text)
```

```
soup = BeautifulSoup(page.content,"html.parser")
```

```
print(soup.prettify())
```

```
data = soup.find_all("div",{"class":"card rteleft facultycard"})
faculty_url = []
for i in range(len(data)):
  divs = data[i].find_all('a')
  for div in divs:
    if div.text != '\xa0':
      div = str(div)
      div = div[9:]
      index = div.find('"')
      div = div[:index]
      faculty_url.append(div)
  
```

```
u = "https://www.iiitd.ac.in/subramanyam"

pg = requests.get(u)
print(pg.text)
```

```
sp = BeautifulSoup(pg.content,"html.parser")
print(sp.prettify())
```

```
dt = sp.find_all("div",{"class":"field field-name-field-contact field-type-text-long field-label-above"})
faculty_email = []
for i in range(len(dt)):
  divs = dt[i].find_all('a')
  for div in divs:
    if div.text != '\xa0':
      div = str(div)
      div = div[9:]
      index = div.find('"')
      div = div[:index]
      faculty_email.append(div)
```

```
print(faculty_email)
```

```
datas = sp.find_all("div",{"class":"field-item even"})
faculty_phone = []
for i in range(len(datas)):
  Divs = datas[i].find_all()
  for div in Divs:
    if div.text != '\xa0':
      faculty_phone.append(div)
```
